```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/winequality-red.csv')
data
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
# define a function

def wine_quality(i):
    if(i>=7):
        return(1)
    else:
        return(0)
```

```python
data['QUALITY'] = data['quality'].apply(wine_quality)
print(data)
```

```python
data.head()
```

```python
# drop the column as are less correlated have skew data and having outliers
data = data.drop(["quality"],axis=1)
data
```

```python
data.describe()
```

```python
# as there are some 0 values in the column of citric acid just impute there with there mean
data['citric acid'] = data['citric acid'].replace(0,data['citric acid'] .mean())
```

```python
# seprating the columns into features and target
x= data.drop("QUALITY",axis=1)
y= data["QUALITY"]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled= sc.fit_transform(x)
```

```python
x_scaled.shape[1]
```

```python
vif = pd.DataFrame()
vif["vif"] = [variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif["features"] = x.columns

vif
```

```python
# as the value of vif>5 so there would be multicollinearity so we can drop these columns
```

```python
x=x.drop(["fixed acidity","density"],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=41)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
# check the accuracy
accuracy_score(y_test,y_pred)
```

```python
# to mprove the accuracy lets try hyperparameter using GridsearchCV algo
```

```python
grid_param = {'criterion':['gini','entropy'],'max_depth':range(2,10),'min_samples_leaf':range(4,10),'min_samples_split':range(3,10),'max_leaf_nodes':range(2,10)}
```

```python
grid_search = GridSearchCV(estimator=clf,param_grid=grid_param,cv=5,n_jobs = -1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_parameters = grid_search.best_params_
print(best_parameters)
```

```python
# now changing the parameters to get the best accuracy
# after many iteration the accuracy not increase muvh so now once give a bigger range and checked the accuracy
```

```python
clf=DecisionTreeClassifier(criterion= 'gini', max_depth= 9, min_samples_leaf= 7, min_samples_split= 7,max_leaf_nodes=5)

clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
# checking the accuracy
accuracy_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred_test = lr.predict(x_test)
```

```python
pred_test
```

```python
# model accuracy

accuracy = accuracy_score(y_test,pred_test)
accuracy
```

```python
conf_matrix = confusion_matrix(y_test,pred_test)
conf_matrix
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred_test))
```

```python
# ROC CURVE 

fpr,tpr,thresholds = roc_curve(y_test,pred_test)
```

```python
print('THRESHOLDS=',thresholds)
print('True positive rate=',tpr)
print('FAlse positive rate=',fpr)
```

```python
plt.plot(fpr,tpr,color = 'orange',label='ROC')
plt.plot([0,1],[0,1],color = 'darkblue',linestyle='--')
plt.xlabel('FAlse positive rate')
plt.ylabel('True positive rate')
plt.title('Receiver operating characterstics(ROC)curve')
plt.legend()
plt.show()
```

```python
# lets chk area it is covering(AUC)
auc_score = roc_auc_score(y_test,pred_test)
print(auc_score)
```
